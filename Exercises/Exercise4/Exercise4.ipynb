```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
# Load and look at the data

Data = pd.read_csv('../Teoph.csv')
Data
```

```python
# Plot for each subject

Figure, Axis = plt.subplots(1,1)
for Label, DataFrame in Data.groupby('Subject'):
    DataFrame.plot(x='Time', y='conc', ax=Axis, label=Label,
                   marker='o', fillstyle='none', linestyle='--')
Axis.set_xlabel('Time (hours)')
Axis.set_ylabel('Concentration (mg/l)')
plt.legend(title='Subject', bbox_to_anchor=(1.05,1))
plt.show()
```
